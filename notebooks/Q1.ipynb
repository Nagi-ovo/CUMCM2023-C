```python
import pandas as pd
import sys 
sys.path.append('/home/aistudio/external-libraries')

file_path1 = '/home/aistudio/work/附件3(处理后).xlsx'
file_path2 = '/home/aistudio/work/每日各品类销量及利润分析.xlsx'

df1 = pd.read_excel(file_path1, sheet_name='Sheet1')
df2 = pd.read_excel(file_path2, sheet_name='Sheet1')
```

```python
date_range_df1 = df1['日期'].min(), df1['日期'].max()
date_range_df2 = df2['Unnamed: 0'].min(), df2['Unnamed: 0'].max()

# Check the unique categories in df1
unique_categories_df1 = df1['分类名称'].unique()

date_range_df1, date_range_df2, unique_categories_df1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Extract the columns related to sales and markup pricing for each category
sales_columns = [col for col in df2.columns if '_销量' in col]
markup_columns = [col for col in df2.columns if '_利润率' in col]

# Create a new dataframe for correlation analysis
correlation_df = df2[sales_columns + markup_columns]

# Calculate the correlation matrix
correlation_matrix = correlation_df.corr()

# Plot the heatmap for correlation between sales and markup pricing
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Between Sales and Markup Pricing for Each Category')
plt.show()
```
